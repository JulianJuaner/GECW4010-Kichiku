#For estimate Normal distribution.
#Get intervals, count occurrence in each, then calculate statistic values.
def getIntervalNormal(data):
    global intervalEndPoint, count, statistic
    for i in range(100):
        alpha = math.sqrt(2)*math.sqrt(var)*(erfinv(i/50-1)) + mean
        intervalEndPoint = np.append(intervalEndPoint, alpha)
    for i in range(99):
        count[i] = ((intervalEndPoint[i] <= data) & (data < intervalEndPoint[i+1])).sum()
    count[99] = (intervalEndPoint[99] <= data).sum()
    for i in range(100):
        statistic[i] = ((float(count[i])-100)**2)/(100)
    print('The test statistic for normal distribution is: ', statistic.sum())
#For estimate Exponential distribution.
#Get intervals, count occurrence in each, then calculate statistic values.

def getIntervalExponential(data):
    global intervalEndPoint, count, statistic
    for i in range(100):
        alpha = beta*np.log(1/(1-i*1/100)) + mindata
        intervalEndPoint = np.append(intervalEndPoint, alpha)
    for i in range(99):
        count[i] = ((intervalEndPoint[i] <= data) & (data < intervalEndPoint[i+1])).sum()
    count[99] = (intervalEndPoint[99] <= data).sum()
    for i in range(100):
        statistic[i] = ((float(count[i])-100)**2)/(100)
    print('The test statistic for exponential distribution is: ', statistic.sum())
